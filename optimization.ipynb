```python
import cobra
from os.path import join
#import iRhto1108N_model with TAG as objective
iRhto1108N_model = cobra.io.read_sbml_model(join(r"C:\Users\helen\Desktop\spring2020\Rao lab\first paper\iRhto_memote-master", "iRhto.xml"))
iRhto1108N_model.objective = "PDAGATpc_rm"
```

```python
max_TAG = iRhto1108N_model.optimize().objective_value
print(max_TAG)
```

```python
import pandas as pd
path1 = "C:/Users/helen/Desktop/spring2020/Rao lab/first paper/iRhto_memote-master/supplementary_files/Supplementary Materials 1.xlsx"
xl = pd.ExcelFile(path1)
df1 = xl.parse('Reactions')
print(df1)
```

```python
from cobra.flux_analysis import flux_variability_analysis
#optimal range for all the reactions
optimum_range = flux_variability_analysis(iRhto1108N_model, iRhto1108N_model.reactions)
```

```python
#find essential reactions using single deletion
from cobra.flux_analysis import (
    single_gene_deletion, single_reaction_deletion, double_gene_deletion,
    double_reaction_deletion)
deletions = single_reaction_deletion(iRhto1108N_model, iRhto1108N_model.reactions)
deletions[:10]
```

```python
essential_reactions = []
for i in range(len(iRhto1108N_model.reactions)):
    if deletions["growth"][i] <= max_TAG * 0.001:
        essential_reactions.append(deletions["growth"].axes[0][i])
len(essential_reactions)
```

```python
#identify the system inwhich the essential reactions belong
list(essential_reactions[0])[0]
df1["id"][0]
systems = []
for i in range(len(essential_reactions)):
    index = [n for n, x in enumerate(df1["id"]) if list(essential_reactions[i])[0] in x]
    systems.append(df1["subsystem"][index])
    essential_reactions[i] = list(essential_reactions[i])[0]
```

```python
essential_reactions_with_system = list(zip(essential_reactions, systems))
essential_reactions_with_system
```
